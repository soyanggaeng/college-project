### 미재개봉 영화와 클러스터 중심 간 거리 구하기 - 가장 가까운 걸로 분류

```python
def sqrt(inp):
    result = inp/2
    for i in range(30):
        try:
            result = (result + (inp / result)) / 2
        except:
            result = 0
    return result
```

```python
import numpy
def euclidean(df, array):
    result = []
    for i in range(len(df)):
        arr = numpy.array(df.iloc[i])
        b = 0
        for j in range(len(arr)):
            b = b + (arr[j] - array[j])**2
        tmp = sqrt(b)
        result.append(tmp)
    return result
```

### 라벨링을 앙상블 형태로

```python
from sklearn.cluster import KMeans
from munkres import Munkres
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd


def doing_kmeans(data, n_clusters):
    KM = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=3)
    KM.fit(data)
    return KM.labels_

def translate_clustering(clt, mapper):
    return np.array([ mapper[i] for i in  clt])

def make_cost_matrix(c1, c2):
    x1 = np.unique(c1)
    x2 = np.unique(c2)
    y1 = x1.size
    y2 = x2.size
    assert(y1 == y2 and np.all(x1 == x2))

    P = np.ones([y1, y2])
    for i in range(y1):
        it_i = np.nonzero(c1 == x1[i])[0]
        for j in range(y2):
            it_j = np.nonzero(c2 == x2[j])[0]
            P_ij = np.intersect1d(it_j, it_i)
            P[i,j] =  -P_ij.size
    return P


def generate_kmeans(data, n_clusters, n_ensembles):
    labels = {}
    for i in range(n_ensembles):
        labels[i]=doing_kmeans(data, n_clusters)

    return labels

def arranging_kmeans(labels):
    m=Munkres()
    new_labels={}
    for i in range(len(labels.keys())):
        if i==0:
            new_labels[i]=labels[i]
        else:
            cost_matrix = make_cost_matrix(labels[i], labels[0])
            indexes = m.compute(cost_matrix)
            mapper = {old:new for (old,new) in indexes }
            new_labels[i] = translate_clustering(labels[i], mapper)
            
    return new_labels

def voting(series):
    val = series.value_counts().sort_values(ascending=False)
    return(val.index[0])

def voting_labels(dict_labels):
    labels = pd.DataFrame.from_dict(dict_labels)
    labels = labels.apply(lambda x:voting(x),axis=1)
    return(labels)

def ensemble_kmeans(data, n_clusters, n_ensembles):
    generated_kmeans_labels = generate_kmeans(data=data, n_clusters=n_clusters, n_ensembles=n_ensembles)
    hungarian = arranging_kmeans(generated_kmeans_labels)
    voted_labels = voting_labels(hungarian)
    
    return voted_labels
```
